```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium

# Load data
data = pd.read_csv('/content/Dataset of Air Quality.csv')

# Visualization 1: Bar Chart - Top 10 cities with the highest average pollutant levels
top_cities = data.groupby('city')['pollutant_avg'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_cities.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Cities with Highest Average Pollutant Levels', fontsize=14)
plt.xlabel('City', fontsize=12)
plt.ylabel('Average Pollutant Level', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Visualization 2: Scatter Plot - Relationship between pollutant_min and pollutant_max
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='pollutant_min', y='pollutant_max', hue='pollutant_id', palette='viridis')
plt.title('Pollutant Min vs Max', fontsize=14)
plt.xlabel('Pollutant Min', fontsize=12)
plt.ylabel('Pollutant Max', fontsize=12)
plt.legend(title='Pollutant ID')
plt.tight_layout()
plt.show()

# Visualization 3: # Group data by city and calculate mean pollutant_avg for each city# Plot the line graph
city_trends = data.groupby('city')['pollutant_avg'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
city_trends.plot(kind='line', color='orange', marker='o', linewidth=2, alpha=0.8)
plt.title('Average Pollutant Levels Across Cities', fontsize=14)
plt.xlabel('City', fontsize=12)
plt.ylabel('Average Pollutant Level', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualization 4: Map with Markers - Station locations
map_ = folium.Map(location=[data['latitude'].mean(), data['longitude'].mean()], zoom_start=5)
for _, row in data.iterrows():
    folium.Marker([row['latitude'], row['longitude']], popup=f"Station: {row['station']}").add_to(map_)
map_.save('station_map.html')

# Visualization 5: Heatmap - Intensity of pollutant_avg in various cities
city_avg = data.groupby('city')['pollutant_avg'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.heatmap(city_avg.head(20).to_frame(), annot=True, cmap='coolwarm', cbar_kws={'label': 'Pollutant Avg'})
plt.title('Heatmap of Pollutant Levels by City', fontsize=14)
plt.tight_layout()
plt.show()

# Visualization 6: Histogram - Distribution of pollutant_avg
plt.figure(figsize=(8, 6))
sns.histplot(data['pollutant_avg'], bins=30, kde=True, color='blue')
plt.title('Distribution of Pollutant Average', fontsize=14)
plt.xlabel('Pollutant Avg', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()

# Visualization 7: Pie Chart - Proportions of pollutant IDs
pollutant_counts = data['pollutant_id'].value_counts()
plt.figure(figsize=(8, 8))
pollutant_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colormap='tab20c')
plt.title('Proportion of Pollutant IDs', fontsize=14)
plt.ylabel('')
plt.tight_layout()
plt.show()

# Visualization 8: Box Plot - Spread of pollutant_avg by state
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='state', y='pollutant_avg', palette='Set2')
plt.title('Spread of Pollutant Average by State', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Pollutant Avg', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualization 9: Density Plot - Concentration of pollutant levels
plt.figure(figsize=(8, 6))
sns.kdeplot(data['pollutant_avg'], shade=True, color='red')
plt.title('Density Plot of Pollutant Average', fontsize=14)
plt.xlabel('Pollutant Avg', fontsize=12)
plt.tight_layout()
plt.show()





```
